```python
# 성적처리 프로그램
# 90점 이상이면 A
# 80점 이상이면 B
# 70점 이상이면 C
# 60점 이상이면 D
# 아니면 F

num = 95

if num >= 90 :
    result ='A' 
elif num >=80 :
    result ='B'
elif num >=70 :
    result ='c'
elif num >=60 :
    result ='D'
else : 
    result ='F'
print(f'학점은 {result} 입니다.')
```

```python
# 수학점수 , 영어점수
# 수학점수와 영어점수가 모두 90점 이상이면 '참 잘했어요'
# 수학점수만 90점 이상 이면 좀 더 노력하세요
# 모두 90점 미만이면 많이 노력 하세요.

math = 90
eng = 80

if math >= 90 and eng >=90 :
    result = '참 잘했어요'
elif math >= 90 and eng < 90 :
    result = '좀 더 노력하세요'
else : 
    result = '많이 노력하세요'
print(result)
```

```python
# 숫자맞추기 게임
import random
com = random.randint(1,100)
print(com)

while True : 
    num = int(input('숫자 입력 :'))
    if num == com:
        print('맞춤')
        break
    elif num > com:
        print('크다')
    elif num < com:
        print('작다') 
```

```python
import random
com = random.randint(1,50)
print(com)

for i in range(10):
    num = int(input())
    if num == com:
        print('OK')
        break
    elif num > com:
        print('Random 값 보다 크다')
    elif num < com:
        print('Random 값 보다 작다')
print('탈락')
```
